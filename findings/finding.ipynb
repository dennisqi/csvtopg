```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
csv = '../data/loan_cleaned_all_tb.csv'
```

```python
df = pd.read_csv(csv)
```

```python
# Number of missing grade
df.grade.isnull().sum()
```

```python
# Show number of records for each grade
grade_counts = df.grade.value_counts().sort_index()
sns.barplot(x=grade_counts.index, y=grade_counts.values)
plt.title('Number of records for each grade')
plt.xlabel('Grade')
plt.ylabel('Number of records')
```

```python
# Number of records changed over time
df.issue_d = pd.to_datetime(df.issue_d)
plt.figure(figsize=(10, 8))
df.issue_d.value_counts().plot()
plt.title('Number of records for each month')
plt.xlabel('Month')
plt.ylabel('Number of records')
```

```python
# Number of records changes over time for each grade
plt.figure(figsize=(15, 11))
grades = sorted(df.grade.unique())
for letter_grade in grades:
    records = df.loc[df['grade'] == letter_grade]
    records.issue_d.value_counts().plot()
plt.legend(grades)
plt.title('Number of records for each month for each grade')
plt.xlabel('Month')
plt.ylabel('Number of records')
```

```python
# Average loan amount changes over time for each grade
plt.figure(figsize=(15, 11))
grades = sorted(df.grade.unique())
for letter_grade in grades:
    records = df.loc[df['grade'] == letter_grade]
    records.groupby(['issue_d']).loan_amnt.mean().plot()
plt.legend(grades)
plt.title('Average loan amount for each month')
plt.xlabel('Month')
plt.ylabel('Number of records')
```

```python
# Average loan amount changes over time for each grade
plt.figure(figsize=(15, 11))
grades = ['A', 'B', 'C', 'D', 'E']
for letter_grade in grades:
    records = df.loc[df['grade'] == letter_grade]
    records.groupby(['issue_d']).loan_amnt.mean().plot()
plt.legend(grades)
plt.title('Average loan amount for each month')
plt.xlabel('Month')
plt.ylabel('Number of records')
```
